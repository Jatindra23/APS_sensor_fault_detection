Resampling data is one of the most commonly preferred approaches to the deal with an imbalanced dataset. There are broadly two types of method for this 1>Undersampling 2> Oversampling. In most cases , oversapmling is preferred over undersampling techniques. the reaasn being in understanding we tend to remove instances from data that maay be carrying some important information

SMOTE->Synthetic mynority Oversampling technique 
        SMOTE is an Oversampling technique where the synthetic samples are generated for the minority class.
        Hybridisation technique involve combining both Underdampling and Ovesampling techniques. this is done to optimize the perfoemance of the classifier models for the samples created as part of these techniques.
        It only duplicates the data and it won't add new information.
